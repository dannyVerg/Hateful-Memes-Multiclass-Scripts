```python
#Binary Text Only Hate Types
import pandas as pd
import numpy as np


pred_tect_type=pd.DataFrame()


pred=pd.read_csv("image_only/hate_type/type_image0.csv")


pred["label"] = pred["label"].apply(lambda x:0 if x==1 else "rest")

#generall overview of the predictions 

for elem in range(1,8):

    pred1_path="image_only/hate_type/type_image"+str(elem)+".csv"
    pred1=pd.read_csv(pred1_path)


    pred1["label"] = pred1["label"].apply(lambda x:"rest" if x==0 else elem)
    
    proba_name="proba"+str(elem)
    label_name="label"+str(elem)
    
    pred[proba_name]=pred1["proba"]
    pred[label_name]=pred1["label"]
    

pred['final_proba']=pred['proba']
pred['final_label']=pred['label']


np_df = np.asarray(pred)

for idx, row in enumerate(np_df[:,:]):
    probas=row[1::2].tolist()
    
    max_proba=max(probas)
    
    finel_proba_index = row.tolist().index(max_proba)
    
    finel_label_index=finel_proba_index+1
    
    
    pred.loc[idx, 'final_label']=row[finel_label_index]
    pred.loc[idx, 'final_proba']=row[finel_proba_index]
        

pred

pred.to_csv("image_only/hate_type/binary_vision_type.csv")
```
