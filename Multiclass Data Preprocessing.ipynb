```python
import pandas as pd
mode="hate_category"
#mode="hate_category"


def csv_to_json(csv_file,json_file, mode, upper_limit=0, down_limit=0):
    df = pd.read_csv(csv_file)
    #df = df.drop('done', 1)
    df = df.drop('Unnamed: 0', 1)

    #for Hate Speech Type Training
    if mode=="hate_type":
        df = df.drop('label Protected Category ', 1)
        df = df.rename(columns = {"label hate Speech Type":"label"})

    #for Hate Speech Type Training 
    elif mode=="hate_category":
        df = df.drop('label hate Speech Type', 1)
        df = df.rename(columns = {"label Protected Category ":"label"})
    else: 
        print("not a valid mode")
        return 
    if upper_limit!=0 and down_limit==0:
        df=df.drop(df.index[upper_limit:])
    elif upper_limit==0 and down_limit!=0:
        df=df.drop(df.index[:down_limit])
    elif upper_limit!=0 and down_limit!=0:
        df=df.drop(df.index[upper_limit:down_limit])
        
    #df.to_json(json_file, orient="records", lines="True")
    global df_type
    df_type=df
    print("saved "+json_file+" at current directory")

csv_to_json("classification  - train.csv", "test_type.jsonl", mode)
```
